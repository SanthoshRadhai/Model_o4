```python
# Select all columns that contain 'p_value' in their name
p_value_columns = df.filter(like='p_value')

# Alternatively, using boolean indexing
p_value_columns = df.loc[:, df.columns.str.contains('p_value')]
p_value_columns.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select all 'p_value' columns from the dataframe
p_value_columns = df.filter(like='p_value')

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Plot histograms for each 'p_value' column
for column in p_value_columns.columns:
    sns.histplot(df[column].dropna(), kde=True, label=column, bins=30, alpha=0.6)

plt.title('Distribution of P-Values')
plt.xlabel('P-Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
